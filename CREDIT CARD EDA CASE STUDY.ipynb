```python
#creating a function for plotting graphs
def plotgraph(col_name):

    
    sns.set_style("ticks")
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=col_name, data=app_non,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Representation of {col_name} for Non-Defaulters',fontsize=15)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=40, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(app_non))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=col_name, data=app_def,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Representation of {col_name} for Defaulters',fontsize=15)    
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=40, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(app_def))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()
```

# Bivariate Analysis of numerical variables

```python
# function for scatter plot for continuous variables
def plotbivar(col1,col2):

    sns.set_style("white")
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.scatterplot(x=col1, y=col2,data=app_non,ax=ax1,color="k")
    ax1.set_xlabel(col1)    
    ax1.set_ylabel(col2)
    ax1.set_title(f'{col1} vs {col2} for Non-Defaulters',fontsize=15)
    
    sns.scatterplot(x=col1, y=col2,data=app_def,ax=ax2)
    ax2.set_xlabel(col1)    
    ax2.set_ylabel(col2)
    ax2.set_title(f'{col1} vs {col2} for Defaulters',fontsize=15)
            
    plt.show()
```

# Merging the Current Application  and Previous Appliction data for further analysis of  the data

```python
## Merging the two files to do further  analysis
Merged_Data = pd.merge(app,prev_app, how='inner', on=['SK_ID_CURR'])
```

```python
app.shape
```

```python
Merged_Data.shape #Checking the number of columns and rows in the merged data 
```

```python
Merged_Data.info() #Checking the column info
```

```python
def plot_merged(col_x,col_y):
    sns.set_style("white")
    sns.despine
    MDat = Merged_Data.pivot_table(values='SK_ID_CURR', 
                      index=col_x,
                      columns=col_y,
                      aggfunc='count')
    MDat=MDat.div(MDat.sum(axis=1),axis='rows')*100
    sns.set()
    MDat.plot(kind='bar',stacked=True,figsize=(15,5))
    plt.title(f'Effect Of {col_x} on Loan Approval')
    plt.xlabel(f'{col_x}')
    plt.ylabel(f'{col_y}%')
    plt.show()
```

```python
plot_merged('FLAG_OWN_CAR','NAME_CONTRACT_STATUS')
```

We see that car ownership doesn't have any effect on application approval or rejection. But we saw earlier that the people who has a car has lesser chances of default. The bank can add more weightage to car ownership while approving a loan amount

Target variable (0 - Non Defaulter 1 - Defaulter )

We can see that the people who were approved for a loan earlier, defaulted less often where as people who were refused a loan earlier have higher chances of defaulting.
